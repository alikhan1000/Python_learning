```python
import pandas as pd
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

df
```

```python
df.info()
df.isnull().sum()
```

```python
def add_student(student):
  global df
  new_data = pd.DataFrame([student])
  df = pd.concat([df,new_data],ignore_index=True)
  print("student Added")
```

```python
add_student({
    "Student_ID":4,
    "Name":"Usman",
    "Class":9,
    "Math":85,
    "English":75,
    "Science":80
})

df
```

```python
def update_marks(student_id, subject, marks):

    global df

    df.loc[df["Student_ID"] == student_id, subject] = marks

    print("Marks Updated")
```

```python
update_marks(1,"Math",95)

df
```

```python
def delete_student(student_id):

    global df

    df = df[df["Student_ID"] != student_id]

    print("Student Deleted")
```

```python
delete_student(4)

df
```

```python
subjects = ["Math","English","Science"]

df["Total"] = df[subjects].sum(axis=1)

df["Average"] = df["Total"]/3

df["Percentage"] = df["Average"]

df
```

```python
def assign_grade(avg):

    if avg >= 80:
        return "A"

    elif avg >= 60:
        return "B"

    elif avg >= 40:
        return "C"

    else:
        return "Fail"
```

```python
df["Grade"] = df["Average"].apply(assign_grade)

df
```

```python
top_student = df.loc[df["Percentage"].idxmax()]

top_student
```

```python
weak_students = df[df["Percentage"] < 50]

weak_students
```

```python
df[subjects].mean()
```

```python
df.groupby("Class")["Percentage"].mean()
```

```python
df.to_csv("updated_students.csv", index=False)

print("File Saved")
```

```python
report = f"""
Student Performance Report

Total Students: {len(df)}

Top Student:
{top_student}

Subject Average:
{df[subjects].mean()}
"""


with open("report.txt","w") as file:
    file.write(report)


print("Report Generated")
```
