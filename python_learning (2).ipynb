```python
#Lists in python
marks = [49.1,50,47,47,27]
print("marks=", marks)
print(type(marks))
print(len(marks))
print(marks[0])
print(marks[1])
ali = ["student",12,"dgk"]
#ali= "k"
age = ali[2]
print("age=" , age)
print(ali)
```

```python
#list slicing

print(marks[-5:-1])
print(marks[5:1])
print(marks[1:5])
print(marks[-5:1])
#mutating
marks.append(110)
#decending order
marks.sort(reverse = True)
#aseending order
marks.sort()
print(marks)
```
