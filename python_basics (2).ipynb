```python
# get marks and print total marks and evarage marks
name= input(("write your name :"))
eng= input(("please enter your ENG marks"))
math= input(("please enter your math marks"))
bio= input(("please enter your bio marks"))
total = int(eng)+int(math)+int(bio)
average = (total)/2;
print("name =" , name)
print("total marks are" , total)
print("Everage marks are " , average)
print("program Successful")
```
